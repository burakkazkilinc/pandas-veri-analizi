### Ülkelere ilişkin kişi başı gayri safi yurt içi hasıla verilerini alın:
Birleşmiş Milletler İnsani Gelişmişlik Raporları (Human Deveoplemen Reports) sitesinde, farklı boyutlar altında çeşitli göstergeleri içeren, ülkelerin 1990-2015 yılları arasındaki İnsani Gelişmişlik Verileri (http://hdr.undp.org/en/data) bulunmaktadır. Bu verilerden "Income/composition of resources" boyutu altında yer alan, "Gross domestic product (GDP) per capita" (kişi başı gayri safi yurt içi hasıla) göstergesine ilişkin verileri indirin ("Gross domestic product (GDP) per capita (2011 PPP $).csv"). Bu dosyadan verileri yükleyin ve bir DataFrame oluşturun.

### Veri Temizleme&Düzenleme:
```
1- Kodlamayı, ISO-8859-1 olarak ayarlayın (read_csv fonksiyonu içindeki encoding parametresi ile)
2- Dosyanın ilk satırını atlayın
3- Son satırı silin
4- Boş sütunları (bütün değerleri 'NaN' olan sütunları) silin
5- Herhangi bir verisi boş ('NaN') olan ülkeleri silin
6- Country sütununda bulunan değerlerin sonundaki ekstra boşlukları silin
7- Country sütunundaki ülke isimlerinin bazılarını aşağıdaki şekilde yeniden adlandırın:
    'Korea (Republic of)':'South Korea',
    'Venezuela (Bolivarian Republic of)':'Venezuela',
    'The former Yugoslav Republic of Macedonia':'Macedonia',
    'Hong Kong, China (SAR)':'Hong Kong'
```

````python
import pandas as pd
import numpy as np
hasila = pd.read_csv('Gdp.csv',encoding='ISO-8859-1',skiprows=1) #Birinci satırı skiprows ile atladım
hasila
````

````python
hasila = hasila[:-1] #Son satırı silme işlemi
hasila
````

````python
hasila.dropna(how='all',axis=1,inplace=True) #Null değerlerden oluşan sütünların silinmesi işlemi
hasila
````

````python
hasila.dropna(how='any',axis=0,inplace=True) #Null değerler içeren satırların silinmesi işlemi
hasila
````

````python
def removeSpace(text):   #Ülke isimlerinde ki boşlukların silinmesi için fonksiyon
    text  = ''.join([i for i in text if not i.isdigit()])    
    return text.strip() 
````

````python
hasila.Country = hasila.Country.apply(removeSpace) #Ülke isimlerinden boşlukların silinmesi
hasila
````

````python
di = {'Korea (Republic of)':'South Korea',
    'Venezuela (Bolivarian Republic of)':'Venezuela',
    'The former Yugoslav Republic of Macedonia':'Macedonia',
    'Hong Kong, China (SAR)':'Hong Kong'}
hasila.replace({"Country": di},inplace = True)  #Belirli ülke isimlerinin Yeniden isimlendirilmesi
````

````python
hasila
````

### Ülkelere ilişkin coğrafi verileri alın:
Geonames veri tabanından ülkelere ilişkin coğrafi verileri içeren dosyadan (http://download.geonames.org/export/dump/countryInfo.txt) verileri yükleyin ve başka bir DataFrame daha oluşturun.

### Veri Temizleme&Düzenleme:
```
1- Dosyanın ilk 50 satırını atlayın
2- Son satırı silin
3- '#ISO' sütununun ismini 'CountryCode' olarak değiştirin
4- 'ISO3', 'ISO-Numeric', 'fips', 'EquivalentFipsCode', 'Postal Code Format', 'Postal Code Regex' ve 'geonameid' sütunlarını silin
5- Kıta (Continent) değeri boş ('NaN') olan ülkelerin kıta değerini 'NA' (North America) olarak değiştirin
```

````python
import pandas as pd
import numpy as np
geonames = pd.read_table('http://download.geonames.org/export/dump/countryInfo.txt',skiprows=50)  #İlk 50 satır atlandı
geonames  
````

````python
geonames=geonames[:-1] #Son satırın silinmesi işlemi
geonames
````

````python
geonames.rename(columns={'#ISO': 'CountryCode'}, inplace=True)  #Sütun ismi değiştirnme işlemi
geonames
````

````python
columns = ['ISO3','ISO-Numeric','fips','EquivalentFipsCode','Postal Code Format','Postal Code Regex','geonameid']
geonames.drop(columns, inplace=True, axis=1)  #Belirtilen sütunların tablodan çıkartılması
geonames
````

````python
geonames.Continent.replace(np.nan,'NA',inplace=True)  #Kıta isimlerinde null değerler yerine 'NA' yazılması
geonames
````

### Soru-6 
Her ülke için o ülkenin komşu ülkelerinin ortalama GDP değerlerinin ortalamasını hesaplayarak yeni bir seri oluşturunuz (eğer bir ülkenin komuşusu yoksa kendi GDP ortalamasını alınız) ve komşuları ile en fazla ortalama GDP farkı olan ilk 20 ülkeyi listeleyin.

````python
d = {'CountryCode': birlesim.CountryCode,'Country': birlesim.Country}  #Ülke kodu ve Ülke isimlerinden oluşan yeni bir DataFrame
yeni_seri = pd.DataFrame(data=d)
yeni_seri
````

````python
liste = []
ana.set_index(ana['CountryCode'],inplace=True) #Index değerinin Ülke kodu olarak değişimi
for Code in ana.CountryCode:  #Her ülkenin kodları üzerinden dolaşımı
    if(pd.notnull(Code)): #ülke kodu bulunmayan ülkeler var
        if(pd.notnull(ana.neighbours[Code])):  #Belirlenen ülkenin bir komşusu var ise
            komsu=ana.neighbours[Code].split(',') #Komşu değerlerinin ayrıştırımı
            for k in komsu:
                if (not(k in ana.neighbours)): #Ülke kodu tabloda bulunan hiçbir ülkeye ait değil ise
                    err=True 
                else:
                    err=False
                
            if(err):  #Ülke kodu bozuk komşu var
                liste.append(ana.loc[Code].avgGDP)
                err=False
            else:  #Ülke kodu bozuk komşu yok
                tablo=ana.loc[komsu]
                liste.append(tablo.avgGDP.mean())              
                               
        else:           
            liste.append(ana.loc[Code].avgGDP)
    else:        
        liste.append(0)
yeni_seri['Komsu_avgGDP']=liste
yeni_seri['avgGdp']=ana.avgGDP
yeni_seri
````

````python
yeni_seri['avg_Fark']=yeni_seri['Komsu_avgGDP']-yeni_seri['avgGdp']
yeni_seri.sort_values(by='avg_Fark',ascending=False).head(20)
````

Dünya'da en çok sayıda ülkede konuşulan ilk 10 dili (bir dilin ülkelere göre yerelleşmiş türlerini, aynı dil olarak kabul edin) ve konuşuldukları ülke sayılarını gösteren bir sütun grafik çizin.

````python
def removeDigit(text):
    text  = ''.join([i for i in text if not i.isdigit()])
    ind=text.find('-')
    if ind > -1:
        text= text[:ind]
    return text.strip()  #'-' den sonra aksan beliritildiği için '-' ve sonrasını silme fonksiyonu
````

````python
d = {'Language': list(set(liste))}
konusulan = pd.DataFrame(data=d)
konusulan.set_index(konusulan['Language'],inplace=True)  #Konusulan diller için bir DataFrame oluşturulması
````

````python
konusulan['Freq']=pd.Series(np.concatenate([x.split() for x in liste])).value_counts() #Her dilin konuşulma sıklığınının tabloda yanına eklenmesi
````
